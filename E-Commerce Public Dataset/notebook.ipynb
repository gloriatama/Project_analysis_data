### Cleaning Data

#### Membersihkan data orders_df

```python
orders_df[orders_df.order_approved_at.isna()]
```

Menangani missing value pada table orders_df

```python
orders_df['order_approved_at'] = orders_df['order_approved_at'].fillna(pd.Timestamp('2018-01-01'))
orders_df['order_delivered_carrier_date'] = orders_df['order_delivered_carrier_date'].fillna(pd.Timestamp('2018-01-01'))
orders_df['order_delivered_customer_date'] = orders_df['order_delivered_customer_date'].fillna(pd.Timestamp('2018-01-01'))
```

Menangani kesalahan tipe data pada table orders_df

```python
datetime_columns = ["order_approved_at", "order_delivered_carrier_date", "order_delivered_customer_date", "order_estimated_delivery_date"]

for column in datetime_columns:
    orders_df[column] = pd.to_datetime(orders_df[column], errors='coerce')
```

## Visualization & Explanatory Analysis

### Pertanyaan 1: Apakah ada bulan-bulan tertentu yang mengalami peningkatan atau penurunan dalam jumlah pesanan?

```python
# Ekstrak informasi bulan dan tahun dari kolom waktu
orders_df['order_month'] = orders_df['order_purchase_timestamp'].dt.to_period('M')

# Hitung jumlah pesanan per bulan
monthly_orders = orders_df.groupby('order_month').size()

# Plot distribusi waktu pemesanan
plt.figure(figsize=(10, 6))
monthly_orders.plot(kind='line')
plt.title('Monthly Order Distribution')
plt.xlabel('Month')
plt.ylabel('Number of Orders')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

### Pertanyaan 2: Apakah ada pola waktu tertentu di mana pesanan lebih cenderung memiliki status tertentu?

```python
# Plot hubungan antara status pemesanan dan waktu pemesanan
plt.figure(figsize=(10, 6))
sns.boxplot(x='order_status', y='order_purchase_timestamp', data=orders_df)
plt.title('Relationship between Order Status and Purchase Time')
plt.xlabel('Order Status')
plt.ylabel('Purchase Time')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

### Pertanyaan 3: Bagaimana distribusi status pemesanan secara keseluruhan?

```python
order_status_counts = orders_df['order_status'].value_counts()

# Plot distribusi status pemesanan
plt.figure(figsize=(8, 6))
order_status_counts.plot(kind='bar')
plt.title('Order Status Distribution')
plt.xlabel('Order Status')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
